```python
# Load JSON
import json
 
# Opening JSON file
f = open('final_qa.json',)
 
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
data
```

```python
# Convert to JSONL format
with open('final_qa.jsonl', 'w') as outfile:
    for item in data:
        # Format each QA pair as a conversation
        conversation = {
            "messages": [
                {
                    "role": "user", 
                    "content": f"{item['question']}"
                },
                {
                    "role": "model",
                    "content": item['answer']
                }
            ]
        }
        # Write each conversation as a JSON line
        json.dump(conversation, outfile)
        outfile.write('\n')
```

```python
with open('final_qa.jsonl', 'w') as outfile:
    for item in data:
        # Format each QA pair as a conversation
        conversation = {
            "contents": [
                {
                    "role": "user",
                    "parts": [
                        {
                            "text": item['question']
                        }
                    ]
                },
                {
                    "role": "model", 
                    "parts": [
                        {
                            "text": item['answer']
                        }
                    ]
                }
            ]
        }
        # Write each conversation as a JSON line
        json.dump(conversation, outfile)
        outfile.write('\n')
```
